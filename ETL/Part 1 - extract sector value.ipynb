```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import mode
```

```python
df=pd.read_csv("temp_datalab_records_linkedin_company.csv")
```

```python
def clean_data(df):
    select_feature=['dataset_id','as_of_date','company_name','followers_count','employees_on_platform','industry']
    newdf=df[select_feature]
    replace_amp=lambda x:x.replace("amp; "," ")
    newdf.loc[:,'industry']=newdf.loc[:,'industry'].astype(str).apply(replace_amp) 
    return newdf
```

```python
newdf=clean_data(df)
```

sector definiation

```python
real_estate=['Commercial Real Estate','Real Estate']
utilities=['Utilities']
communications=['Telecommunications','Public Relations and Communications','Wireless','Broadcast Media']
it=["Internet",'Information Technology and Services','Semiconductors','Information Services','Computer & Network Security','Computer Hardware','Computer Networking','Computer Software']
financial=['Investment Banking','Banking','Insurance','Capital Markets','Venture Capital & Private Equity','Investment Management','Financial Services']
healthcare=['Hospital & Health Care','Pharmaceuticals','Health, Wellness and Fitness','Medical Devices']
consumer_staples=['Tobacco','Consumer Goods','Consumer Electronics','Food & Beverages','Food Production','Wine and Spirits','Supermarkets','Furniture','Dairy','Wholesale']
consumer_discretionary=['Retail','Apparel & Fashion','Restaurants','Computer Games','Sporting Goods','Leisure, Travel & Tourism','Hospitality','Luxury Goods & Jewelry','Gambling & Casinos','Fine Art','Graphic Design',
       'Photography','Entertainment']
basic_material=['Mining & Metals','Chemicals','Building Materials','Packaging and Containers','Paper & Forest Products','Plastics','Glass, Ceramics & Concrete']
engery=['Oil & Energy']
industrials=['Electrical/Electronic Manufacturing','Automotive','Mechanical or Industrial Engineering','Defense & Space','Machinery','Aviation & Aerospace','Construction','Industrial Automation','Railroad Manufacture']
```

```python
def clean_get_sector(newdf,sector_list):
    company_list=[]
    for i in sector_list:
        company_list.extend(list(newdf[newdf['industry']==i]['company_name'].unique()))
    
    sector_df=pd.DataFrame(columns=['dataset_id','as_of_date','company_name','followers_count','employees_on_platform','industry'])
    for i in company_list:
       
        temp_company_df=pd.DataFrame(columns=['dataset_id','as_of_date','company_name','followers_count','employees_on_platform','industry'])
        temp_company_df=newdf[newdf['company_name']==i]
        temp_company_df['industry']=mode(temp_company_df['industry'])
        #newdf[newdf['company_name']=="Twitter"].replace({"industry":"nan"},mode(newdf[newdf['company_name']=="Twitter"].loc[:,'industry']))
        temp_company_df.drop_duplicates(inplace=True)
        if temp_company_df.shape[0]>1000:
            sector_df=pd.concat([sector_df, temp_company_df])
    
    sector_df.set_index(pd.DatetimeIndex(sector_df['as_of_date']),inplace=True)
    return sector_df
        
```

```python
real_estate_sector_df=clean_get_sector(newdf,real_estate)
real_estate_sector_df.to_csv('real_estate_sector_df.csv',index=False)

utilities_sector_df=clean_get_sector(newdf,utilities)
utilities_sector_df.to_csv('utilities_sector_df.csv',index=False)

communications_sector_df=clean_get_sector(newdf,communications)
communications_sector_df.to_csv('communications_sector_df.csv',index=False)

it_sector_df=clean_get_sector(newdf,it)
it_sector_df.to_csv('it_sector_df.csv',index=False)

financial_sector_df=clean_get_sector(newdf,financial)
financial_sector_df.to_csv('financial_sector_df.csv',index=False)

healthcare_sector_df=clean_get_sector(newdf,healthcare)
healthcare_sector_df.to_csv('healthcare_sector_df.csv',index=False)

consumer_staples_sector_df=clean_get_sector(newdf,consumer_staples)
consumer_staples_sector_df.to_csv('consumer_staples_sector_df.csv',index=False)

consumer_discretionary_sector_df=clean_get_sector(newdf,consumer_discretionary)
consumer_discretionary_sector_df.to_csv('consumer_discretionary_sector_df.csv',index=False)

basic_material_sector_df=clean_get_sector(newdf,basic_material)
basic_material_sector_df.to_csv('basic_material_sector_df.csv',index=False)

engery_sector_df=clean_get_sector(newdf,engery)
engery_sector_df.to_csv('engery_sector_df.csv',index=False)

industrials_sector_df=clean_get_sector(newdf,industrials)
industrials_sector_df.to_csv('industrials_sector_df.csv',index=False)
```
